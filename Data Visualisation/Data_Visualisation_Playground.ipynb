## 1.1 - Matplotlib

Below is the most basic way of plotting by just using `matplotlib`

```python
plt.plot(fifa) 
plt.show();
```

```python
# Implementation of matplotlib function
import matplotlib.pyplot as plt
import numpy as np
  
fig = plt.figure()
x = np.arange(20) / 50
y = (x + 0.1)*2
  
val1 = [True, False] * 10
val2 = [False, True] * 10
  
plt.errorbar(x, y, 
             xerr = 0.1, 
             xlolims = True, 
             label ='Line 1')
  
y = (x + 0.3)*3
  
plt.errorbar(x + 0.6, y, 
             xerr = 0.1,
             xuplims = val1,
             xlolims = val2,
             label ='Line 2')
  
y = (x + 0.6)*4
plt.errorbar(x + 1.2, y,
             xerr = 0.1, 
             xuplims = True,
             label ='Line 3')
  
plt.legend()
  
fig.suptitle('matplotlib.pyplot.show() Example 1')
plt.show()
```

```python
# Implementation of matplotlib function
import matplotlib.pyplot as plt
import numpy as np
  
x = np.linspace(0, 10, 500)
y = np.sin(x**2)+np.cos(x)
  
fig, ax = plt.subplots()
  
ax.plot(x, y, label ='Line 1')
  
ax.plot(x, y - 0.6, label ='Line 2')
  
ax.legend()
  
fig.suptitle('matplotlib.pyplot.show() Example 2')
plt.show()
```

# 4 - Scatter Plots (Relationship)

Scatter plots are good to observe correlation between variables, still one should not neglect that corellation does not imply causality!!!.

```python
insurance_data = pd.read_csv('Data/insurance.csv')
insurance_data.head()
```

```python
candy_data = pd.read_csv('Data/candy.csv', index_col='id')
candy_data.head()
```

The dataset contains 83 rows, where each corresponds to a different candy bar. There are 13 columns:
- `competitorname` contains the name of the candy bar.
- the next 9 columns (from `chocolate` to `pluribus`) describe the candy. For instance, rows with chocolate candies have "Yes" in the 'chocolate' column (and candies without chocolate have "No" in the same column).
- 'sugarpercent' provides some indication of the amount of sugar, where higher values signify higher sugar content.
- 'pricepercent' shows the price per unit, relative to the other candies in the dataset.
- 'winpercent' is calculated from the survey results; higher values indicate that the candy was more popular with survey respondents.


```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges']);
```

```python
sns.scatterplot(data = candy_data, x='pricepercent', y='winpercent');
```

## 4.1 - Adding a regression line

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges']);
```

```python
sns.regplot(data = candy_data, x='sugarpercent', y='winpercent');
```

Since the regression line has a slightly positive slope, this tells us that there is a slightly positive correlation between winpercent and sugarpercent. Thus, people have a slight preference for candies containing relatively more sugar.

## 4.2 - Colour-coded scatter plots

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
sns.scatterplot(data = candy_data, x='pricepercent', y='winpercent', hue='chocolate');
```

## 4.3 - Colour code and regression line

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data);
```

```python
sns.lmplot(data = candy_data, x='pricepercent',y='winpercent', hue='chocolate')
```

I'll begin with the regression line for chocolate candies. Since this line has a slightly positive slope, I can say that more expensive chocolate candies tend to be more popular (than relatively cheaper chocolate candies). Likewise, since the regression line for candies without chocolate has a negative slope, I can say that if candies don't contain chocolate, they tend to be more popular when they are cheaper. One important note, however, is that the dataset is quite small -- so I shouldn't invest too much trust in these patterns! To inspire more confidence in the results, I should add more candies to the dataset.
